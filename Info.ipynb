```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))


files = [1,2,4,6,8,10,12]
for i in files:
    sizes = []
    times = []
    with open(f"consumed_time {i}.txt", "r") as file:
        next(file)
        for line in file:
            parts = line.split()
            sizes.append(int(parts[2]))
            times.append(float(parts[5]))
    plt.plot(sizes, times, marker='o', label=f': {i}')

plt.xlabel('Size')
plt.ylabel('Time, sec')
plt.title('Time vs Size')
plt.grid()
plt.legend()
plt.show()
```
